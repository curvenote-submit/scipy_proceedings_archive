```python
from IPython.display import display,SVG
```

```python
import figurefirst as fifi
layout = fifi.FigureLayout('additional_features_layout.svg')
layout.make_mplfigures()
#SVG items - Panels A & B
x = np.linspace(0,2*pi,20)
animal_1 = sin(x)
animal_2 = arctan(x)
cdict1 = {'r1':0.3,'r2':0.1,'r3':0.9}
cdict2 = {'path1':0.3,'path2':0.1,'path3':0.9}
#change box fills
for key,patch in layout.svgitems['svggroup'].items():
    clev = cdict1[key]
    hexi = matplotlib.colors.rgb2hex(plt.cm.viridis(clev))
    patch.style['fill'] = str(hexi)
#change path fills
for key,patch in layout.svgitems['pathgroup'].items():
    clev = cdict2[key]
    hexi = matplotlib.colors.rgb2hex(plt.cm.viridis(clev))
    patch.style['fill'] = str(hexi)
#change the text and text color
layout.svgitems['l1'].style['fill'] = str(hexi)
layout.svgitems['l1'].text = str(clev)
layout.apply_svg_attrs()

#Path specs - Panel C & D
layout.axes['path_spec_ax'].plot(x,animal_1,**layout.pathspecs['animal_1'].mplkwargs())
layout.axes['path_spec_ax'].plot(x,animal_2,**layout.pathspecs['animal_2'].mplkwargs())
layout.axes['path_spec_ax'].axvspan(pi*0.5,pi*1.5,**layout.pathspecs['trial_epoch'].mplkwargs())
    
layout.save('additional_features.svg')
close('all')
display(SVG('additional_features.svg'))
```
